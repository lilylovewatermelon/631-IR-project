```python
import time
import pandas as pd
from lyricsgenius import Genius
import requests


songs = []
# num = 500
tracks_df = pd.read_csv('combined_spotify_songs_2000_2022.csv');
# index = track_df[track_df['song_name'] == 'Take Care'].index
# start = index +1

tracks_df = tracks_df[20000:24000]
# len(tracks_df)

def search_song_with_retry(song_name, artist_name, retries=3, delay=1):
    """Attempt to fetch song lyrics with retries."""
    for attempt in range(retries):
        try:
            song = genius.search_song(song_name, artist_name)
            return song
        except requests.exceptions.Timeout:
            print(f"Timeout occurred for '{song_name}' by '{artist_name}'. Attempt {attempt + 1} of {retries}.")
            time.sleep(delay)
        except requests.exceptions.RequestException as e:
            print(f"Request exception for '{song_name}' by '{artist_name}': {e}")
            return None
    return None

for index, row in tracks_df.iterrows():
    # if index > num:
    #     break
    song_name = row['song_name']
    artist_name = row['artist_names']
    album_id = row['album_id']
    song_id = row['song_id']

    if "Noise" in artist_name:
        continue

    song = search_song_with_retry(song_name, artist_name.split(",")[0])

    if song:
        lyrics = song.lyrics
        song_info = {
            'id': song_id,
            'song_name': song_name,
            'artist_names': artist_name,
            'album_id': album_id,
            'lyrics': lyrics
        }
        songs.append(song_info)
    else:
        print(f"No lyrics found for {song_name} by {artist_name}")

songs_df2 = pd.DataFrame(songs)
songs_df2.to_csv('songs_4000.csv', index=False)
```
